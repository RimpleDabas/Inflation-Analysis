```python
import pandas as pd
import numpy as np
import calendar
```

```python
df = pd.read_csv('Retail Canada.csv')
df.head(10)
```

```python
df.columns
```

```python
df = df[['REF_DATE', 'GEO', 'Products', 'VALUE','UOM' ]]
df.head(10)
```

```python
df.isnull().values.any()
```

```python
df= df.rename(columns = {'GEO' : 'Location'})
df.dtypes
```

```python
df['Products Name'] = df['Products'].astype('string').str.split(',').str[0]

df['Amount'] = df['Products'].astype('string').str.split(',').str[1]

df.sample(10)
```

```python
df['Quantity'] = df['Amount'].astype('string').str.replace('per', '1')
df.sample(10)
```

```python
df['Year'] = df['REF_DATE'].astype('string').str.split('-').str[0]
df['Month'] = df['REF_DATE'].astype('string').str.split('-').str[1]
df.sample(10)
```

```python
df_check = df.loc[df['Products Name']=='Milk']
df_check.sample(10)
```

```python
df['Month'] = df['Month'].astype(int).apply(lambda x: calendar.month_abbr[x])

```
